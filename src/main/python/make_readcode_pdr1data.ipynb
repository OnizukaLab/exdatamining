```python
def extract_attribute(sql_statement):
    cast = set()

    for i in sql_statement.split(','):
        temp = i.split(' ')[1].replace('\n', '')
        if temp not in cast:
            cast.add(temp)
    return cast
```

```python
def make_class(sql_statement):
    attr_list = list()
    for i in sql_statement.split(','):
        attr_list.append(
            ','
            +
            i.replace('\n','').replace(' ', ' : ').
            replace(' bigint', ' BigInt').replace(' smallint', ' Int').
            replace(' double', ' Double').replace(' string', ' String').
            replace(' float', ' Float').replace(' int', ' Int').replace(' boolean', ' Boolean')
        )
    return attr_list
```

```python
def make_scala_statement(sql_statement):
    attribute_list = list()
    j = 0
    for i in sql_statement.split(','):
        attribute_list.append(
            'val ' 
            +
            i.replace('\n','').replace(' ', ' = elms(%s).to').replace('%s', str(j)).
            replace('smallint', 'int').replace('bigint', 'BigInt').
            replace('double', 'Double').replace('string', 'String').
            replace('int', 'Int').replace('float', 'Float').replace('boolean', 'Boolean')
        )
        j += 1
        if attribute_list[-1] == "val coord = elms(3).toString":
            j += 2
        
    return attribute_list
```
